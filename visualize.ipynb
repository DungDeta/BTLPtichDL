```python
# 1. Tách thành phố và quận/huyện từ địa chỉ
def extract_city_district(address_str):
    if pd.isna(address_str):
        return None, None
    parts = [part.strip() for part in address_str.split(',')]
    if not parts:
        return None, None
    city = parts[-1]
    district = None
    if city == "Hà Nội":
        if len(parts) >= 2:
            district = parts[-2]
    return city, district

data[['Extracted City', 'Extracted District']] = data['Địa chỉ'].apply(lambda x: pd.Series(extract_city_district(x)))
data_hanoi_mask = data['Extracted City'] == 'Hà Nội'
data = data[data_hanoi_mask].copy()
print(f"Số lượng bất động sản ở Hà Nội: {len(data)}")

data.rename(columns={'Extracted District': 'Quận/Huyện'}, inplace=True)
if 'Extracted City' in data.columns:
    data.drop(columns=['Extracted City'], inplace=True)

def standardize_district(district_name):
    if pd.isna(district_name):
        return district_name
    # Loại bỏ tiền tố "Quận" hoặc "Huyện"
    if district_name.startswith('Quận '):
        return district_name.replace('Quận ', '', 1)
    elif district_name.startswith('Huyện '):
        return district_name.replace('Huyện ','', 1)
    return district_name

data['Quận/Huyện'] = data['Quận/Huyện'].apply(standardize_district)
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y=data['Quận/Huyện'], order=data['Quận/Huyện'].value_counts().head(20).index)
plt.title('Số lượng tin đăng theo Quận/Huyện ở Hà Nội')
plt.xlabel('Số lượng')
plt.ylabel('Quận/Huyện')
plt.show()
```

```python
# Violin plot: So sánh phân phối giá theo quận/huyện
district_order = data['Quận/Huyện'].value_counts().head(20).index
data_top_districts = data[data['Quận/Huyện'].isin(district_order)]

plt.figure(figsize=(14, 8))
sns.violinplot(x='Quận/Huyện', y=(data_top_districts['Mức giá (VNĐ)']/1_000_000),
               data=data_top_districts, palette='Set3', inner='quartile',
               order=district_order)
plt.title('So sánh phân phối giá theo Quận/Huyện (triệu VNĐ)')
plt.xlabel('Quận/Huyện')
plt.ylabel('Mức giá (triệu VNĐ)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

data_valid = data.dropna(subset=['Mức giá (VNĐ)', 'Diện tích (m2)'])
district_price = data_valid.groupby('Quận/Huyện')['Mức giá (VNĐ)'].agg(['mean', 'count'])
district_area = data_valid.groupby('Quận/Huyện')['Diện tích (m2)'].agg(['mean'])
district_stats = pd.DataFrame({
    'price_mean': district_price['mean'],
    'price_count': district_price['count'],
    'area_mean': district_area['mean']
}).sort_values('price_mean', ascending=False).head(20)

# Chỉ lấy các quận/huyện có ít nhất 10 mẫu để kết quả đáng tin cậy hơn
district_stats = district_stats[district_stats['price_count'] >= 10]

# Tạo cột mới với giá trị là triệu VNĐ để dễ hiển thị
district_stats['price_million'] = district_stats['price_mean'] / 1_000_000
fig, ax1 = plt.subplots(figsize=(14, 8))
color = 'tab:blue'
ax1.set_xlabel('Quận/Huyện')
ax1.set_ylabel('Mức giá trung bình (triệu VNĐ)', color=color)
# Tạo danh sách nhãn tên quận kèm số lượng mẫu
x_labels = [f"{idx}\n(n={district_stats['price_count'][idx]:.0f})" for idx in district_stats.index]
x_pos = np.arange(len(district_stats.index))

# Vẽ cột với vị trí x được chỉ định rõ ràng
bars = ax1.bar(x_pos, district_stats['price_million'], color=color, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)

# Đặt nhãn trục x bằng tên quận và số lượng mẫu
ax1.set_xticks(x_pos)
ax1.set_xticklabels(x_labels, rotation=45, ha='right')

# Hiển thị giá trị trên đầu mỗi cột
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
            f'{height:.1f}', ha='center', va='bottom', color=color, fontweight='bold')

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Diện tích trung bình (m2)', color=color)
ax2.plot(x_pos, district_stats['area_mean'], 'o-', color=color, linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

# Hiển thị giá trị diện tích trên đường line
for i, (x, y) in enumerate(zip(x_pos, district_stats['area_mean'])):
    ax2.annotate(f'{y:.1f}', (x, y), textcoords="offset points",
                xytext=(0,5), ha='center', color=color, fontweight='bold')

plt.title('Mức giá trung bình và diện tích trung bình theo Quận/Huyện')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Biểu đồ kde 2D: Mật độ phân bố theo giá và diện tích
plt.figure(figsize=(10, 8))
sns.kdeplot(x='Diện tích (m2)', y=(data['Mức giá (VNĐ)']/1_000_000),
           data=data, cmap="viridis", fill=True, thresh=0, levels=20)
plt.title('Mật độ phân bố theo giá và diện tích')
plt.xlabel('Diện tích (m2)')
plt.ylabel('Mức giá (triệu VNĐ)')
plt.tight_layout()
plt.show()
```

```python
# Biểu đồ cột nhóm: So sánh số phòng ngủ với giá trung bình
bedroom_price = data.groupby('Số phòng ngủ (phòng)')['Mức giá (VNĐ)'].mean().reset_index()
bedroom_price = bedroom_price[bedroom_price['Số phòng ngủ (phòng)'] <= 6]  # Lọc các giá trị hợp lý
bedroom_price['Mức giá (triệu)'] = bedroom_price['Mức giá (VNĐ)'] / 1_000_000

plt.figure(figsize=(10, 6))
bars = plt.bar(bedroom_price['Số phòng ngủ (phòng)'], bedroom_price['Mức giá (triệu)'], color='skyblue')

# Thêm số giá lên đầu mỗi cột
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
            f'{height:.1f}', ha='center', va='bottom')

plt.title('Giá trung bình theo số phòng ngủ')
plt.xlabel('Số phòng ngủ (phòng)')
plt.ylabel('Mức giá trung bình (triệu VNĐ)')
plt.xticks(bedroom_price['Số phòng ngủ (phòng)'])
plt.tight_layout()
plt.show()
```

```python
# Histogram: Phân bố số lượng phòng ngủ
plt.figure(figsize=(10, 6))
rooms_data = data['Số phòng ngủ (phòng)'].dropna()
# Đếm số lượng mỗi giá trị
room_counts = rooms_data.value_counts().sort_index()
print("Phân bố số phòng ngủ:")
for rooms, count in room_counts.items():
    print(f"{int(rooms) if rooms.is_integer() else rooms} phòng ngủ: {count} BĐS ({count/len(rooms_data)*100:.1f}%)")

# Vẽ histogram với giá trị rõ ràng trên trục x
bars = plt.bar(room_counts.index, room_counts.values, color='skyblue')
plt.title('Phân bố theo số lượng phòng ngủ')
plt.xlabel('Số phòng ngủ (phòng)')
plt.ylabel('Số lượng BĐS')
plt.xticks(room_counts.index)
# Thêm số liệu lên đầu mỗi cột
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{height:.0f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# Histogram: Phân bố số lượng phòng tắm
plt.figure(figsize=(10, 6))
bathroom_data = data['Số phòng tắm (phòng)'].dropna()
# Đếm số lượng mỗi giá trị
bathroom_counts = bathroom_data.value_counts().sort_index()
print("Phân bố số phòng tắm:")
for rooms, count in bathroom_counts.items():
    print(f"{int(rooms) if rooms.is_integer() else rooms} phòng tắm: {count} BĐS ({count/len(bathroom_data)*100:.1f}%)")

# Vẽ histogram với giá trị rõ ràng trên trục x
bars = plt.bar(bathroom_counts.index, bathroom_counts.values, color='lightgreen')
plt.title('Phân bố theo số lượng phòng tắm')
plt.xlabel('Số phòng tắm (phòng)')
plt.ylabel('Số lượng BĐS')
plt.xticks(bathroom_counts.index)
# Thêm số liệu lên đầu mỗi cột
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{height:.0f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
# Biểu đồ phân tích tương quan giữa số phòng ngủ và số phòng tắm
plt.figure(figsize=(12, 8))
rooms_correlation = data.dropna(subset=['Số phòng ngủ (phòng)', 'Số phòng tắm (phòng)'])
# Chỉ hiển thị các giá trị phổ biến để tránh quá tải biểu đồ
rooms_correlation = rooms_correlation[(rooms_correlation['Số phòng ngủ (phòng)'] <= 6) &
                                      (rooms_correlation['Số phòng tắm (phòng)'] <= 6)]

heatmap_data = pd.crosstab(rooms_correlation['Số phòng ngủ (phòng)'],
                          rooms_correlation['Số phòng tắm (phòng)'])

# Vẽ heatmap hiển thị mối tương quan giữa số phòng ngủ và phòng tắm
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Mối tương quan giữa số phòng ngủ và số phòng tắm')
plt.xlabel('Số phòng tắm (phòng)')
plt.ylabel('Số phòng ngủ (phòng)')
plt.tight_layout()
plt.show()
```
